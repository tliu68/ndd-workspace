```python
%pylab inline
from sklearn.ensemble import RandomForestRegressor
```

```python
#Dataset creation
x = np.arange(256)
x = x[:,newaxis]
y = np.sin(2*np.pi*x/128)
n = 0.5*np.random.randn(256)
n = n[:,newaxis]
y = y+n

z = np.arange(min(x), max(x),0.5)
z = z[:,newaxis]

#Fitting Model
reg = RandomForestRegressor(n_estimators=200, min_samples_split=3, random_state=0) 
reg.fit(x,y)

#Visualization
plt.figure(figsize=(16,7))
plt.scatter(x, y, alpha=0.3, c='red')
plt.plot(z, reg.predict(z))
plt.legend(['RF Regression','Observation'])
```
